```
import urllib.request
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
```

```
init_url = "http://www.amc.seoul.kr/asan/healthinfo/disease/diseaseList.do?"
```

```
source = "http://www.amc.seoul.kr/asan/healthinfo/disease/diseaseList.do?diseaseKindId=C0000"
```

```
req = urllib.request.urlopen(source)
res = req.read()

soup = BeautifulSoup(res, 'html.parser')
sub_source = soup.find_all('li', class_='listCon')
```

```
urls = []
for s_source in sub_source:
    if "diseaseKindId" in (s_source.a['href']):
        _url = s_source.a['href'][-21:]
        _menu = re.sub('<.+?>', "", str(s_source))
        urls.append((_url, _menu.strip()))
urls
```

```
df = pd.DataFrame(columns=['class', 'disease', 'syn'])
row_cnt = 0
df
```

```
row_cnt = 0
for idx, url in tqdm(urls):
    disease_classification = url[1]
    # page 개수 count
    req = urllib.request.urlopen(init_url + url[0])
    res = req.read()
    soup = BeautifulSoup(res, 'html.parser')
    links = soup.select('div.pagingWrapSec > span.numPagingSec > a > span')
    page_count = int(links[-1].text) + 1
    
    for idx in range(1, page_count):
        req = urllib.request.urlopen(init_url + "pageIndex={}&".format(idx) + url[0])
        res = req.read()
        soup = BeautifulSoup(res, 'html.parser')
        links = soup.select('div.listCont > ul.descBox > li')
        for link in links:
            if '동의어' in str(link):
                disease = link.find_all('a')[0].text
                syn = link.find_all('dd')[-1].text.strip()
                df.loc[row_cnt, 'class'] = url[1]
                df.loc[row_cnt, 'disease'] = disease
                df.loc[row_cnt, 'syn'] = syn
                row_cnt += 1
```

```
output_path = "/content/drive/My Drive/Colab Notebooks/output/"
df.to_csv(output_path + "disease_list.csv", index=False)
```
