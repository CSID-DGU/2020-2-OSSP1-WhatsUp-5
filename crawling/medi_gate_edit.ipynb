```python
import csv
import re
```

```python
f = open('medi_gate.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medi_gate_edit.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

for line in rdr:
    try:
        wr.writerow([line[0],line[1], line[2],line[3]])
    except IndexError:
        print("에러발생")
        
f.close()
w.close()
```

```python
f = open('medi_gate_edit.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medi_gate_edit1.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

count = 1
for line in rdr:
    try:
        newline = line[3]
        start = newline.find("이 기사를")
        if(start != -1):
            end = newline.find("데이터 축적이 충분치 않습니다.")
            if(end != -1):
                print(line[0],line[1], "기사 유형제거")
                newline = newline[:start] + newline[end:]
        wr.writerow([line[0],line[1],line[2], newline])
        count += 1
    except IndexError:
        print("에러발생")
    
f.close()
w.close()
```

```python
f = open('medi_gate_edit1.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medi_gate_edit2.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

for line in rdr:
    newline = line[3]
    start = newline.find("데이터 축적이 충분치 않습니다.")
    if(start != -1):
        newline = newline.replace("데이터 축적이 충분치 않습니다.", "")
        print(line[0], line[1], "기사 유형제거")
    wr.writerow([line[0], line[1], line[2], newline])
f.close()
w.close()
```

```python
f = open('medi_gate_edit2.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medi_gate_edit3.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

for line in rdr:
    newline = line[3]
    newline = newline.replace('\"', "").replace("▲","").replace("ⓒ메디게이트뉴스","").replace("사진","").replace('"','').replace('“',"").replace("자료실 바로가기",'')
    newline = newline.replace('게티이미지 뱅크', '').replace("이 기사를 많이 읽고 있습니다",'')
    name = line[1]
    name = name.replace('"', "")
    wr.writerow([line[0], name, line[2], newline])
f.close()
w.close()
```

```python
f = open('medi_gate_edit3.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medi_gate_edit4.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

for line in rdr: 
    newline = line[3]
    while True:            
            start = newline.find("(")
            end = newline.find(")")
            if(start == -1):
                break;
            if start < end:
                print(line[0], line[1], "제거완료")
                newline = newline[:start] + newline[end+1:]      
            else:
                print(start, end)
                break;
    wr.writerow([line[0],line[1],line[2],newline])
f.close()
w.close()
```
