```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import WebDriverException

from bs4 import BeautifulSoup
import time
import csv
```

```python
#삼성병원의 질환백과의 정보를 크롤링함
site = 'https://kcdcode.kr/browse/contents/0'

#csv 파일 저장을 위한 f, utf-8을 인코딩으로 사용
f = open('cdc.csv', 'w',encoding='utf-8')
#파일 객체를 생성
wr = csv.writer(f)

row = 1
listnum = 1
cleft_name = ["","","","","","",""]
```

```python
def cleft(depth, row):
    num = 1
    for i in range(depth, 6):
        cleft_name[i] = ""
    cleft_last = driver.find_element_by_css_selector("div#kcdf_cleft" + str(depth) + " ul.kcdf_cname li:last-child a").text
    while True:
        cleft_selector = driver.find_element_by_css_selector("div#kcdf_cleft" + str(depth) + " ul.kcdf_cname li:nth-child("+ str(num)+") a")
        cleft_name[depth] = cleft_selector.text
        driver.execute_script("arguments[0].scrollIntoView(true);", cleft_selector)
        cleft_selector.send_keys('\n')
        
        try:
            cleft_next = driver.find_element_by_css_selector("div#kcdf_cleft" + str(depth+1) + " ul.kcdf_cname li:last-child a")
            row = cleft(depth+1, row)                                            
        except WebDriverException:
            row = write_csv(row)
        if cleft_last == cleft_selector.text:
            break
        num +=1
    return row
```

```python
def write_csv(row):
    driver.implicitly_wait(3)
    try:
        term = driver.find_element_by_css_selector("div#kcdf_cright div#kcdf_cright_cts ul").text
    except WebDriverException:
        term = ""    
    wr.writerow([row, cleft_name[0],cleft_name[1],cleft_name[2],cleft_name[3],cleft_name[4],cleft_name[5],cleft_name[6],term])
    print("출력중" + str(row))
    row += 1   
    return row
```

```python
#현재 PC에 설치한 chromedriver의 위치를 잡아준다.
chrome_driver = 'D:/study/Python/SW/chromedriver.exe'

#chrome으로 driver를 실행
driver = webdriver.Chrome(chrome_driver)

#조건 성립 시간을 위한 wait 객체 설정
wait = WebDriverWait(driver,20)

#driver의 브라우저 크기를 최대화
driver.maximize_window()
#최대화를 하지 않았을 경우 마우스 포인터가 정확히 div를 클릭하지 못하여 크롤링을 못할 수 있음

#웹의 컴포넌트 로딩을 기다리기 위한 implicitly_wait 함수, 로딩이 끝나면 바로 진행된다.
driver.implicitly_wait(3)

#미리 설정해둔 삼성서울병원의 질환백과 사이트로 이동한다.
driver.get(site)



cdc_list_last = driver.find_element_by_css_selector("div#kcdf_cleft ul.kcdf_cname li:last-child").text
while True:
    cdc_list_selector = driver.find_element_by_css_selector("div#kcdf_cleft ul.kcdf_cname li:nth-child("+str(listnum)+") a")
    cleft_name[0] = cdc_list_selector.text
    driver.execute_script("arguments[0].scrollIntoView(true);", cdc_list_selector)
    cdc_list_selector.send_keys('\n')
    row = cleft(1, row)
    if cleft_name[0] == cdc_list_last:
        break;
    listnum +=1
```
