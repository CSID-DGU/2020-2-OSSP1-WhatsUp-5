```python
import csv
import re
```

```python
f = open('g_health_edit.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('g_health_word_processing.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

count = 1
for line in rdr:
    try:
        newline = line[0]
        start = newline.find("［")
        end = newline.find("］")
        if start < end :
            print(str(count) + "제거완료")
            newline = newline[:start] + newline[end+1:]
        wr.writerow([count, newline])
        count += 1
    except IndexError:
        print("에러발생")
        
    
f.close()
w.close()
```

```python
f = open('g_health_word_processing.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('g_health_word_processing2.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)


count = 1
for line in rdr:
    try:
        newline = line[1]
        while True:            
            start = newline.find("<")
            end = newline.find(">")
            if(start == -1):
                break;
            if start < end:
                print(str(count) + "제거 완료")
                newline = newline[:start] + newline[end+1:]      
            else:
                print(start, end)
                break;
        wr.writerow([count, newline])
        count += 1
    except IndexError:
        print("에러발생")
        
    
f.close()
w.close()
```
