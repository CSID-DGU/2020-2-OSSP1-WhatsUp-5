```python
import csv
import re
```

```python
f = open('medical_times.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medical_times_edit.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

count = 1
for line in rdr:
    try:
        newline = line[3]
        start = newline.find("학술학회")
        if(start != -1):
            print(line[0],line[1], "학술학회제거")
            newline = newline[:start]
        wr.writerow([line[0],line[1],line[2], newline])
        count += 1
    except IndexError:
        print("에러발생")
        
    
f.close()
w.close()
```

```python
f = open('medical_times_edit.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medical_times_edit2.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

for line in rdr:
    newline = line[3]
    newline.replace('"', "").replace("자료사진","")
    while True:            
            start = newline.find("(")
            end = newline.find(")")
            if(start == -1):
                break;
            if start < end:
                print(line[0], line[1], "제거완료")
                newline = newline[:start] + newline[end+1:]      
            else:
                print(start, end)
                break;
    wr.writerow([line[0],line[1],line[2],newline])
    
f.close()
w.close()
```

```python
f = open('medical_times_edit2.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medical_times_edit3.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

for line in rdr:
    newline = line[3]
    newline.replace('""', "").replace("▲", "")
    while True:
        start = newline.find("|")
        end = newline.find("|", start + 1)
        if(start == -1):
            break;
        if start < end:
            print(line[0], line[1], "제거완료")
            newline = newline[:start] + newline[end+1:]      
        else:
            print(start, end)
            break;
    wr.writerow([line[0],line[1],line[2],newline])
f.close()
w.close()
```

```python
f = open('medical_times_edit3.csv', 'r', encoding = 'utf-8')
rdr = csv.reader(f)

w = open('medical_times_edit4.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(w)

for line in rdr:
    newline = line[3]
    newline = newline.replace('\"', "").replace("▲","")
    name = line[1]
    name = name.replace('"', "")
    wr.writerow([line[0], name, line[2], newline])
f.close()
w.close()
```
