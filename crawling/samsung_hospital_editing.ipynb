```python
import pandas
import numpy


data = pandas.read_csv("samsungHospital_edit.csv")
datalist = list(numpy.array(data.values.tolist()))

edited_data = []

def text_mining(before_list):
    dash = 3
    text = ""
    for a in range(len(before_list)):
        if((before_list[a].find('●')!= -1) and (before_list[a].find('다.')== -1)):
            text += (before_list[a]+' ')
            dash = 0
        elif(before_list[a].startswith('-') and dash == 0):
            add_d = before_list[a]+ ' '
            text += add_d
            dash = 1
        elif(before_list[a].startswith('-') and dash == 1):
            add_d = ' '+before_list[a]+' '
            text += add_d
            dash = 1
        elif((before_list[a].find('●')== -1) and dash == 0):
            text += (': '+ before_list[a]+' ')
            dash = 1
        else:
            text += before_list[a]+' '               
    return text

def editing(i):
    disease = ""
    definition = ""
    cause = ""
    symptom = ""
    diagnosis = ""
    treatment = ""
    complications = ""
    
    #질병명 추출
    disease = datalist[i][2]
    
    #질병 정의 추출
    definition = datalist[i][3]
    if(definition != ''):
        definition = definition.replace('\n', ' ')
    
    #발병 이유 추출
    cause = datalist[i][4]
    if(cause != ''):
        cause = cause.replace('\n', ' ')
    
    #질병 증상 추출
    symptoms = datalist[i][5]
    if(symptoms != ''):
        symp = datalist[i][5].split('\n')
        symptom = text_mining(symp)
    else:
        symptom = symptoms
        
    symptom = symptom.replace('●', '● ')
    symptom = symptom.replace('.●', '. ●')
    symptom = symptom.replace(' . ', '')
    symptom = symptom.replace('  ', ' ')
    
    #질병 진단 추출
    diagno = datalist[i][6]
    if(diagno != ''):
        dia = datalist[i][6].split('\n')
        diagnosis = text_mining(dia)
    else:
        diagnosis = diagno
        
    diagnosis = diagnosis.replace('●', '● ')
    diagnosis = diagnosis.replace('.●', '. ●')
    diagnosis = diagnosis.replace(' . ', '')
    diagnosis = diagnosis.replace('  ', ' ')
    
    #질병 치료법 추출
    treat = datalist[i][7]
    if(treat != ''):
        t = datalist[i][7].split('\n')
        treatment = text_mining(t)
    else:
        treatment = treat
        
    treatment = treatment.replace('●', '● ')
    treatment = treatment.replace('.●', '. ●')
    treatment = treatment.replace(' . ', '')
    treatment = treatment.replace('  ', ' ')
    
    #질병 합병증 추출
    comp = datalist[i][8]
    if(comp != ''):
        com = datalist[i][7].split('\n')
        complications = text_mining(com)
    else:
        complications = comp
        
    complications = complications.replace('●', '● ')
    complications = complications.replace('.●', '. ●')
    complications = complications.replace(' . ', '')
    complications = complications.replace('  ', ' ')
    
        
    edited_data.append([disease, definition, cause, symptom, diagnosis, treatment, complications])

    
    

for i in range(len(datalist)):
    editing(i)
    
print(len(edited_data))
```

```python
dataframe = pandas.DataFrame(edited_data)
dataframe.to_csv("C:/Users/lg/python/samsung_hospital_db.csv", header=False, index=False)
print('finish')
```
